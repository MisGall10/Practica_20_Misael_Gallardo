# Ejercicio 1: Normalización Min-Max

1.	Cargar el dataset wine de sklearn.
2.	Seleccionar una de las características numéricas (por ejemplo, alcohol).
3.	Realizar una prueba de normalidad (usando scipy.stats.shapiro) sobre los datos originales.
4.	Aplicar la normalización Min-Max a los datos seleccionados.
5.	Realizar la prueba de normalidad nuevamente sobre los datos normalizados.
6.	Graficar los datos originales y los normalizados utilizando matplotlib o seaborn.
7.	Responder las preguntas siguientes.

Preguntas:
1. ¿Cómo afectó la normalización Min-Max a la distribución de los datos?

No afecto en nada porque este metodo de normalizacion mantiene la distribucion de los datos ya que no altera la relacion en los valores.

2. ¿Qué interpretas del p-valor antes y después de la normalización?

Que confirmamos que la distribucion no cambio porque no mejoro en nada el p-valor.


```python
#Importamos las librerias a utilizar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
#Cargamos el dataset de wine
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
#Seleccionamos la columna 'alcohol'
alcohol = df['alcohol']
#Realizamos las pruebas de normalidad usando spicy.stats.shapiro
print(f"\nPruebas de normalidad para la columna alcohol")
# Prueba de Shapiro-Wilk
shapiro_test = stats.shapiro(alcohol)
print(f"Shapiro-Wilk Test: p-valor = {shapiro_test.pvalue:.5f}")
#Aplicamos la normalizacion Min-Max a los datos seleccionados.
scaler = MinMaxScaler(feature_range=(0, 1))
alcohol_scaled = scaler.fit_transform(alcohol.values.reshape(-1, 1))
#Volvemos a realizar la prueba de normalidad con los datos transformados
print(f"\nPruebas de normalidad para la columna alcohol transformada")
#Prueba de Shapiro-Wilk
shapiro_test_escaled = stats.shapiro(alcohol_scaled)
print(f"Shapiro-Wilk Test: p-valor = {shapiro_test_escaled.pvalue:.5f}")
#Visualizar los datos originales vs normalizados
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.scatter(range(len(alcohol)), alcohol, color='red', label="Original")
plt.title("Datos Originales 'alcohol'")

plt.subplot(1, 2, 2)
plt.scatter(range(len(alcohol_scaled)), alcohol_scaled, color='blue', label="Min-Max Scaling")
plt.title("Datos Normalizados 'alcohol' (0,1)")

plt.show()
```

# Ejercicio 2: Normalización Z-Score

Instrucciones:
1.	Cargar el dataset wine de sklearn.
2.	Seleccionar una de las características numéricas (por ejemplo, malic_acid).
3.	Realizar una prueba de normalidad (usando scipy.stats.shapiro) sobre los datos originales.
4.	Aplicar la normalización Z-Score a los datos seleccionados.
5.	Realizar la prueba de normalidad nuevamente sobre los datos normalizados.
6.	Graficar los datos originales y los normalizados utilizando matplotlib o seaborn.
7.	Responder las preguntas siguientes.

Preguntas:
1.	¿Qué diferencia observas entre la distribución original y la normalizada con Z-Score?

La escala de las variables dependientes, es lo unico que cambia, ya que al no tener los datos una forma de distribucion normal, la distribucion sigue siendo no normal

2.	¿Qué implicaciones tiene la normalización Z-Score en el análisis de los datos?

Esta normalizacion solo cambia la escala de las variables, pero no cambia la forma de la distribucion si los datos no siguen una distribucion normal


```python
#Importamos las librerias a utilizar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from scipy import stats
#Cargamos el dataset de wine
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
#Seleccionamos la columna 'malic_acid'
malic_acid = df['malic_acid']
#Realizamos las pruebas de normalidad usando spicy.stats.shapiro
print(f"\nPruebas de normalidad para la columna malic_acid")
# Prueba de Shapiro-Wilk
shapiro_test = stats.shapiro(malic_acid)
print(f"Shapiro-Wilk Test: p-valor = {shapiro_test.pvalue:.5f}")
#Aplicamos la normalizacion Z-score a los datos seleccionados.
scaler = StandardScaler()
malic_acid_scaled = scaler.fit_transform(malic_acid.values.reshape(-1, 1))
#Volvemos a realizar la prueba de normalidad con los datos transformados
print(f"\nPruebas de normalidad para la columna malic_acid transformada")
#Prueba de Shapiro-Wilk
shapiro_test_escaled = stats.shapiro(malic_acid_scaled)
print(f"Shapiro-Wilk Test: p-valor = {shapiro_test_escaled.pvalue:.5f}")
#Visualizar los datos originales vs normalizados
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.scatter(range(len(malic_acid)), malic_acid, color='red', label="Original")
plt.title("Datos Originales 'malic_acid'")

plt.subplot(1, 2, 2)
plt.scatter(range(len(malic_acid_scaled)), malic_acid_scaled, color='blue', label="Min-Max Scaling")
plt.title("Datos Normalizados 'malic_acid' (0,1)")

plt.show()
```

# Ejercicio 3: Transformación Logarítmica
Instrucciones:
1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar una transformación logarítmica (np.log) a los datos.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación logarítmica a la distribución de los datos?

Al aplicar la transformacion logratimica a datos que vienen dados de manera exponecial, ademas de notar que la nueva distribucion se comporta de maneta lineal, lo cual nos inidica una dependencia linean en ellas.

2.	¿Qué interpretas del p-valor antes y después de la transformación logarítmica?

El pvalor = 0 nos dice que los datos no siguen una distribucion normal., en cambio el pvalor = 0.0221 nos inidica que los datos transformados aun no siguen una distribucion normal, porque pvalor aun es lo suficientemente pequeño como para no seguir perfectamente la distribucion normal.



```python
#Importamos las bibliotecas a usar
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
#Cargamos un conjjunto de datos aleatorios generado con numpy
x = np.linspace(1, 10, 100)
y = np.exp(0.5 * x) + np.random.randint(0, 5, size=len(x))
#Aplicamos la transformacion logaritmica
y_log = np.log(y)
#Realizamos las pruebas de normalidad usando spicy.stats.shapiro
print(f"\nPruebas de normalidad para los datos")
# Prueba de Shapiro-Wilk
shapiro_test = stats.shapiro(y)
print(f"Shapiro-Wilk Test: p-valor = {shapiro_test.pvalue:.5f}")
#Volvemos a realizar la prueba de normalidad con los datos transformados
print(f"\nPruebas de normalidad para los datos transformada")
#Prueba de Shapiro-Wilk
shapiro_test_escaled = stats.shapiro(y_log)
print(f"Shapiro-Wilk Test: p-valor = {shapiro_test_escaled.pvalue:.5f}")
# Graficamos los datos antes y después de la transformación
plt.figure(figsize=(12, 6))
# Graficamos los datos originales
plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue', alpha=0.7)
plt.title("Datos Originales")
plt.xlabel("x")
plt.ylabel("y")
# Graficamos los datos transformados
plt.subplot(1, 2, 2)
plt.scatter(x, y_log, color='green', alpha=0.7)
plt.title("Datos con Transformación Logarítmica")
plt.xlabel("x")
plt.ylabel("log(y)")

plt.tight_layout()
plt.show()
```

# Ejercicio 4: Transformación Raíz Cuadrada
Instrucciones:
1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar una transformación de raíz cuadrada a los datos.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación raíz cuadrada a la distribución de los datos?

Practicamente trato de aplicar una distribucion normal en base a los datos, obteniendo un mejor pvalue para, lo cual nos dice que la grafica tiende a seguir una distribucion normal, pero no lo suficiente como para considerar que si lo es

2.	¿Qué efectos tiene la raíz cuadrada sobre los valores extremos de los datos?

Los reduce siguiendo la tendencia de la raiz cuadrada de sus valores extremos.


```python
#Importamos la bibliotecas a utilizar
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
#Cargamos un conjunt ode datos aleatorios generados con numpy
x = np.random.randint(1, 100, size=100)
#Aplicamos una transformacion de raiz cuadrad a los datos
x_sqrt = np.sqrt(x)
#Realizamos las pruebas de normalidad usando spicy.stats.shapiro
print(f"\nPruebas de normalidad para los datos")
# Prueba de Shapiro-Wilk
shapiro_test = stats.shapiro(x)
print(f"Shapiro-Wilk Test: p-valor = {shapiro_test.pvalue:.5f}")
#Volvemos a realizar la prueba de normalidad con los datos transformados
print(f"\nPruebas de normalidad para los datos transformada")
#Prueba de Shapiro-Wilk
shapiro_test_escaled = stats.shapiro(x_sqrt)
print(f"Shapiro-Wilk Test: p-valor = {shapiro_test_escaled.pvalue:.5f}")
# Gráfico antes de la transformación
plt.figure(figsize=(8, 4))
sns.histplot(x, bins=10, kde=True, color='red')
plt.title("Distribución Original")
plt.show()
# Gráfico después de la transformación
plt.figure(figsize=(8, 4))
sns.histplot(x_sqrt, bins=10, kde=True, color='blue')
plt.title("Distribución después de Transformación Raíz Cuadrada")
plt.show()
```

# Ejercicio 5: Transformación Box-Cox
Instrucciones:
1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar la transformación Box-Cox a los datos utilizando scipy.stats.boxcox.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación Box-Cox a la distribución de los datos?

Paso de tener un considerable sesgo hacia la derecha, a tener una distribucion aparentemente normal. Para confirmar ello hacemos una prueba de normalidad y notamos que por su valor pvalue aun no cumple con lo necesario para considerarla normal

2.	¿Qué interpretación tiene el valor lambda obtenido en la transformación Box-Cox?

Es el valor que tiende a transformar los datos dinamicamente en base a cada valor, lo que mejora la estabilidad y el rendimiento en modelos de ML


```python
#Importamos las librerias a usar
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import boxcox
#Cargamos un conjunto de datos aleatorios generados con numpy
x = np.random.exponential(scale=2, size=100)
#Aplicación de la Transformación Box-Cox
x_transformed, lambda_opt = boxcox(x)
#Realizamos las pruebas de normalidad usando spicy.stats.shapiro
print(f"\nPruebas de normalidad para los datos")
# Prueba de Shapiro-Wilk
shapiro_test = stats.shapiro(x)
print(f"Shapiro-Wilk Test: p-valor = {shapiro_test.pvalue:.5f}")
#Volvemos a realizar la prueba de normalidad con los datos transformados
print(f"\nPruebas de normalidad para los datos transformada")
#Prueba de Shapiro-Wilk
shapiro_test_escaled = stats.shapiro(x_sqrt)
print(f"Shapiro-Wilk Test: p-valor = {shapiro_test_escaled.pvalue:.5f}")
# Gráfico de los datos originales
plt.figure(figsize=(8, 4))
sns.histplot(x, kde=True, bins=30, color='red')
plt.title("Distribución Original (Sesgada)")
plt.show()
# Gráfico de los datos transformados
plt.figure(figsize=(8, 4))
sns.histplot(x_transformed, kde=True, bins=30, color='blue')
plt.title(f"Distribución Transformada con Box-Cox (λ óptimo = {lambda_opt:.2f})")
plt.show()
```
